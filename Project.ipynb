```python
import pandas as pd
import glob
import os
import numpy as np
```

```python
file = os.path.join(r"C:\Users\Niang\Desktop\M1IA\projet semestre\Notebook\dataset\data_with_season", "*.csv")
```

```python
files = glob.glob(file)
```

```python
files
```

```python
#concat les data csv en un seul dataframe.
pd.set_option('display.max_columns', None)
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
```

```python
df
```

```python
df['year'] = df['Date'].apply(lambda x: x[6:10])
df['month'] = df['Date'].apply(lambda x: x[3:5])
```

```python
df = df.drop('Date', axis=1)
```

```python
df
```

```python
df.head(5)
```

```python
df["FTR"] = np.where(df["FTR"]=="H", 'H', 'NH')
df["HTR"] = np.where(df["HTR"]=="H", 'H', 'NH')
```

```python
df
```

```python
df["HomeTeam"] = df["HomeTeam"].astype("category").cat.codes
df["AwayTeam"] = df["AwayTeam"].astype("category").cat.codes
```

```python
#df= df.drop(["B365H", "B365D","BWH","BWD", "BWA","IWH", "IWD", "IWA", "PSH", "PSD", "PSA","WHH" ], axis=1)
```

```python
df = df.drop(columns=df.columns[22:123])
```

```python
#df
```

```python
df = df.drop(["Div"], axis=1)
df = df.drop(["FTHG"], axis=1)
df = df.drop(["FTAG"], axis=1)
```

```python
df
```

```python
df.loc[df['FTR'] == 'NH', 'FTR'] = 0
df.loc[df['FTR'] == 'H', 'FTR'] = 1

df.loc[df['HTR'] == 'NH', 'HTR'] = 0
df.loc[df['HTR'] == 'H', 'HTR'] = 1
```

```python
from sklearn.preprocessing import MaxAbsScaler

abs_scaler = MaxAbsScaler()
abs_scaler.fit(df)

abs_scaler.max_abs_

scaled_df = abs_scaler.transform(df)
df_scaled = pd.DataFrame(scaled_df, columns=df.columns)
```

```python
df_scaled
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
      Dense(20, input_shape=(20,), activation='relu'),
      Dense(10, activation='relu'),
      Dense(2, activation='sigmoid')
])
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model.compile(
    Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
y = df_scaled.loc[:, 'FTR']
X = df_scaled.drop('FTR', axis=1)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=67)
```

```python
X_train
```

```python
history = model.fit(
    x=X_train, 
    y=y_train, 
    validation_split=0.8, 
    batch_size=10, 
    epochs=30, 
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau()], 
    shuffle=True, 
    verbose=2
)
```

```python
y_pred = model.evaluate(X_test, y_test, batch_size=10)
```
