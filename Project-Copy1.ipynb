```python
df = df.drop(["Div"], axis=1)
df = df.drop(["FTHG"], axis=1)
df = df.drop(["FTAG"], axis=1)
```

```python
df
```

```python
df.loc[df['FTR'] == 'NH', 'FTR'] = 0
df.loc[df['FTR'] == 'H', 'FTR'] = 1

df.loc[df['HTR'] == 'NH', 'HTR'] = 0
df.loc[df['HTR'] == 'H', 'HTR'] = 1
```

```python
from sklearn.preprocessing import MaxAbsScaler

abs_scaler = MaxAbsScaler() #une instance de l'objet MaxAbsScaler
abs_scaler.fit(df)  # ajustée aux données d'entrée

abs_scaler.max_abs_   #La valeur maximale absolue de chaque colonne est stockée

scaled_df = abs_scaler.transform(df) #mise à jour des données
df_scaled = pd.DataFrame(scaled_df, columns=df.columns)  #convertion en un nouveau dataframe
```

```python
df_scaled
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
      Dense(20, input_shape=(20,), activation='relu'),
      Dense(10, activation='relu'),
      Dense(8, activation='sigmoid')
])
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model.compile(
    Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
y = df_scaled.loc[:, 'FTR']
X = df_scaled.drop('FTR', axis=1)
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=67)
```

```python
X_train
```

```python
history = model.fit(
    x=X_train, 
    y=y_train, 
    validation_split=0.8, 
    batch_size=10, 
    epochs=20, 
    callbacks=[tf.keras.callbacks.ReduceLROnPlateau()], 
    shuffle=True, 
    verbose=2
)
```

```python
y_pred = model.evaluate(X_test, y_test, batch_size=10)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label="Training accuracy")
plt.plot(history.history['val_accuracy'], label="Validation accuracy")
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import seaborn as sns 
df_history = pd.DataFrame(history.history)
my_plot = sns.lineplot(data=df_history[["loss","val_loss"]])
my_plot.set_xlabel('Epochs')
my_plot.set_ylabel('Loss')
plt.legend(labels=["Training", "Validation"])
plt.title('Training and Validation Loss')
plt.show()
```
